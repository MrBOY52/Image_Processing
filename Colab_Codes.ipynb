```python
import cv2
from ultralytics import YOLO
import random

# --- GÃœNCELLENMÄ°Å AYARLAR ---
# 1. IÅŸÄ±k/Levha Modelinin Yolu (Ä°lk eÄŸittiÄŸin model)
path_model_light = '/content/runs/detect/trafik_modelim/weights/best.pt'

# 2. AraÃ§ Modelinin Yolu (ÅÄ°MDÄ° EÄÄ°TTÄ°ÄÄ°N MODEL)
# EÄŸer eÄŸitim hata verdiyse veya dosya yoksa yolu kontrol et.
path_model_vehicle = '/content/runs/detect/arac_modeli/weights/best.pt'

# 3. Video YollarÄ±
input_video = '/content/test_video.mp4'
output_video = '/content/iki_ozel_model_birlesik.mp4'
# -----------------------------

print("Ã–zel modeller yÃ¼kleniyor...")
# Hata yÃ¶netimi: Dosyalar var mÄ±?
try:
    model_lights = YOLO(path_model_light)
    model_vehicles = YOLO(path_model_vehicle)
    print("Modeller baÅŸarÄ±yla yÃ¼klendi.")
except Exception as e:
    print(f"HATA: Model dosyalarÄ± bulunamadÄ±. LÃ¼tfen eÄŸitimlerin bittiÄŸinden emin ol.\nDetay: {e}")
    # Kodun geri kalanÄ± Ã§alÄ±ÅŸmasÄ±n diye durdurabiliriz ama devam edelim, belki biri vardÄ±r.

# Video AyarlarÄ±
cap = cv2.VideoCapture(input_video)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

# Renkler
COLOR_LIGHTS = (0, 255, 0)    # YeÅŸil
COLOR_VEHICLES = (255, 0, 0)  # Mavi (FarklÄ± olsun)

print("Video iÅŸleniyor...")

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # --- MODEL 1: IÅIKLAR ---
    results_lights = model_lights.predict(frame, conf=0.4, verbose=False)

    # --- MODEL 2: ARAÃ‡LAR ---
    # ArtÄ±k 'classes=...' filtresi kullanmÄ±yoruz Ã§Ã¼nkÃ¼ senin modelin sadece araÃ§larÄ± Ã¶ÄŸrendi.
    results_vehicles = model_vehicles.predict(frame, conf=0.5, verbose=False)

    # --- Ã‡Ä°ZÄ°M 1: ARAÃ‡LAR (Mavi) ---
    for r in results_vehicles:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])

            # SÄ±nÄ±f adÄ±nÄ± al
            cls_id = int(box.cls[0])
            label = model_vehicles.names[cls_id]

            cv2.rectangle(frame, (x1, y1), (x2, y2), COLOR_VEHICLES, 2)
            cv2.putText(frame, f"{label} {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, COLOR_VEHICLES, 2)

    # --- Ã‡Ä°ZÄ°M 2: IÅIKLAR (YeÅŸil) ---
    for r in results_lights:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])

            cls_id = int(box.cls[0])
            label = model_lights.names[cls_id]

            cv2.rectangle(frame, (x1, y1), (x2, y2), COLOR_LIGHTS, 2)
            cv2.putText(frame, f"{label} {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, COLOR_LIGHTS, 2)

    out.write(frame)

cap.release()
out.release()
print(f"Ä°ÅŸlem bitti! Video: {output_video}")
```
